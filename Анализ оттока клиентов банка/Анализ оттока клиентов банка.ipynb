# Описание проекта

**Банки — Анализ оттока клиентов**

**Цель исследования:
Проанализировать клиентов регионального банка и выделить сегменты клиентов,
которые склонны уходить из банка.**

**Задачи**
- Провестиисследовательский анализ данных: определить все значимые признаки отточности (интервалы значений характеристик, которые связаны с повышенным оттоком, сравнить портреты типичных клиентов, которые склонны и не склонныуходить из банка)

- Сформулировать и проверить статистические гипотезы.
    - *Проверьте гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались.*
    - *Проверьте гипотезу различия баланса между теми клиентами, которые ушли и теми, которые остались.*
    - *Проверьте гипотезу различия скоринга между теми клиентами, которые ушли и теми, которые остались.*
    
- Объединяя признаки отточности, сформировать сегменты, отбобрать из них лучшие и дать по ним рекомендации 

**Описание датасетов**

Датасет содержит данные о клиентах банка «Метанпром». 

Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.

```python
#создадим функцию по объему каждой категории внутри признака 
def histiogram_categ(df1, df2, column_x, column_name):
    fig = go.Figure()
    fig.add_trace(go.Histogram(
                           x=df1[column_x], histnorm='percent', marker_color=colors[0],
                name='данные с пропусками в балансе'))
    fig.add_trace(go.Histogram(
                           x=df2[column_x], histnorm='percent', marker_color=colors[1],
                name='данные без пропусков в балансе'))
    fig.update_traces(marker_line_color = 'gray',
                  marker_line_width = 2, opacity = 1)
    fig.update_layout(
    {'title':{'text':'Доля категорий в общем объеме параметра <br> {}'.format(column_name), 
              'x': 0.5, 'y': 0.95, 'font_size': 20}},
    legend_orientation="h", 
    legend=dict(x=0.2, y=-0.2),#положение легенды
    autosize=False,
    width=900,
    height=500,
    plot_bgcolor='white',
    )
    fig.update_xaxes(title_text=column_name, showgrid=True, gridcolor='silver')
    fig.update_yaxes(title_text='доля в общем объеме(%)', showgrid=True, gridcolor='silver', dtick=5)

    #fig.show() #показать диаграмму
    fig.show("png") #показать диаграмму для Github
```

**Далее рассмотрим распределение по непрерывным характеристикам.**

```python
#создадим функцию
def histiogram_quantitative(df1, df2, column_x, column_name):
    fig = go.Figure()
    fig.add_trace(go.Histogram(
                           x=df1[column_x], histnorm='density', marker_color=colors[0], 
                name='данные с пропусками в балансе'))
    fig.add_trace(go.Histogram(
                           x=df2[column_x], histnorm='density', marker_color=colors[1], 
                name='данные без пропусков в балансе'))
    fig.update_traces(marker_line_color = 'gray',
                  marker_line_width = 2, opacity = 1)
    fig.update_layout(
    {'title':{'text':'Гистограмма распределения параметра <br> {}'.format(column_name), 
              'x': 0.5, 'y': 0.95, 'font_size': 20}},
    barmode='stack',
    legend_orientation="h", 
    legend=dict(x=0.2, y=-0.2),#положение легенды
    autosize=False,
    width=900,
    height=500,
    plot_bgcolor='white',
    )
    fig.update_traces(marker_line_color = 'gray', marker_line_width = 1, opacity = 0.7)
    fig.update_xaxes(title_text=column_name, showgrid=True, gridcolor='silver')
    fig.update_yaxes(title_text='частота', showgrid=True, gridcolor='silver')

    #fig.show() #показать диаграмму
    fig.show("png") #показать диаграмму для Github
```

```python
histiogram_quantitative(balance_nan, balance_not_nan, 'age', 'возраст')
```

```python
histiogram_quantitative(balance_nan, balance_not_nan, 'score', 'скоринг')
```

```python
histiogram_quantitative(balance_nan, balance_not_nan, 'salary', 'зарплата')
```

Среди признаков с непрерывными данными параметры 'age' и 'salary' имеют достаточно похожее распределение.

Со скорингом все наоборот, основная масса данных с пропусками в балансе приходится на скоринг ниже 800 либо выше 870. В диапазоне 800-870 имеется огромный правал в распределении. В то время как в данных без пропусков на этот диапазон приходится значительная часть данных. А вот количество данных с низким скорингом крайне мало.

**Таким образом можно выделить, следующие признаки клиентов, у которых имеются пропущенные значения в балансе:'**
- credit card' - наличие кредитной карты в 82% случаев,
- 'equity' - 95%  клиентов с самым низком количество балов собственности-0,

Пропуски вызваны особенностями отражения баланса(учитываются только собственные средства клиентов-текущие счета и депозиты). А так как не все клиенты банка имеют данные продукты,  то соответственно счетов просто нет, отсюда пропуски.

```python
#создадим функцию длясоздания сводной таблицы и диаграммы по категориальным признакам
def pivot_categ_bar(df, column, column_name, binary):
#сводная таблица с количеством не отточных клиентов по выбранному параметру
    churn_0 = df.query('churn == False').groupby([column])['user_id'].count().reset_index()
    churn_0.columns = [column_name, 'не отточные'] #переименуем столбцы

#сводная таблица с количеством отточных клиентов по выбранному параметру
    churn_1 = data.query('churn == True').groupby([column])['user_id'].count().reset_index()
    churn_1.columns = [column_name, 'отточные'] #переименуем столбцы

#сводная таблица с общим количеством клиентов по каждой категории
    churn_all = data.groupby([column])['user_id'].count().reset_index()
    churn_all.columns = [column_name, 'всего'] #переименуем столбцы

#соединяем все таблицы 
    churn_all = churn_all.merge(churn_0, on=column_name).merge(churn_1, on=column_name)
#считаем доли отточных и не отточных в категориях    
    churn_all['доля не отточные'] = round((churn_all['не отточные'] / churn_all['всего'])*100, 2)
    churn_all['доля отточные'] = round((churn_all['отточные'] / churn_all['всего'])*100, 2)
#создадим доп параметр для бинарных признаков - при True нумирация заменяется на строчные значения    
    if binary == True:
        churn_all[column_name] = churn_all[column_name].replace({1: 'да', 0: 'нет'})
#считаем доли отточных и не отточных в общем объеме признака
    churn_all['отточные доля в общем объеме'] = (churn_all['отточные']/churn_all['всего'].sum()).apply('{:.2%}'.format)
    churn_all['не отточные доля в общем объеме'] = (churn_all['не отточные']/churn_all['всего'].sum()).apply('{:.2%}'.format)
#выводим общую своднуб таблицу
    display(churn_all)
#создадимдиаграммы с количеством клиентов, а также долей по каждой категории в признаке    
    fig = make_subplots(rows=1, cols=2, #количество строк и столбцов
                    subplot_titles=['Количество в категории', 'Доля в объеме категории(в %)'], #названия диаграмм по порядку 
                    shared_yaxes=True, #общая ось у
                    horizontal_spacing=0.06 #пространство между диаграммами 
                   )
#первая диаграмма
    fig.add_trace(go.Bar(name='не отточные', y = churn_all[column_name], x = churn_all['не отточные'],
                         text=churn_all['не отточные'], hovertext=churn_all['не отточные доля в общем объеме'], #доп текст
                         orientation='h', marker_color=colors[0]),1,1)
    fig.add_trace(go.Bar(name='отточные', y = churn_all[column_name], x = churn_all['отточные'], 
                         text=churn_all['отточные'], hovertext=churn_all['отточные доля в общем объеме'], #доп текст 
                         orientation='h', marker_color=colors[1]),1,1)
#вторая диаграмма
    fig.add_trace(go.Bar(name='не отточные', y = churn_all[column_name], x = churn_all['доля не отточные'],
                         text=churn_all['доля не отточные'], orientation='h', marker_color=colors[0]),1,2)
    fig.add_trace(go.Bar(name='отточные', y = churn_all[column_name], x = churn_all['доля отточные'],
                         text=churn_all['доля отточные'], orientation='h', marker_color=colors[1]),1,2)
#зададим параметры диаграмм
    fig.update_layout(
        {'title':{'text':'Распределение отточных и не отточных клиентов по параметру <br> {}.'.format(column_name), 
                  'x': 0.5, 'y': 0.95, 'font_size': 20}}, #название общей диаграммы, расположение текста, размер текста
        barmode='stack', #вид диаграмм
        legend_orientation="h", 
        legend=dict(x=0.2, y=-0.2),#положение легенды
        autosize=False,
        width=900,
        height=500,
        paper_bgcolor='white',
        plot_bgcolor='white', #фоновый цвет диаграмм
        uniformtext_minsize=8, #размер шрифта данных
    )
    fig.update_annotations(font_size = 16)
    fig.update_xaxes(title_text='значение') #ось х
    fig.update_yaxes(title_text=column_name) #ось у
#выводим получившуюся диаграмму
    #fig.show() #показать диаграмму
    fig.show("png") #показать диаграмму для Github
```

```python
#создадим функция для постороения гистограммы или kde-графика
def histiogram_plotly(df, column_x, column_name):
#отфильтруем 2 таблицы по значению churn
    df_churn_0 = df[df.churn == False]
    df_churn_1 = df[df.churn == True]
#передадим датафреймы в список для построения kde-графика
    hist_data = [df_churn_1[column_x], df_churn_0[column_x]]
    group_labels = ['отточные', 'не отточные']
#заменим бинарные значения на строчные
    color = df['churn'].replace({1: 'отточные', 0: 'не отточные'})
#строим диаграмму    
    fig = go.Figure()
#kde-график
    fig = ff.create_distplot(hist_data, group_labels, histnorm='probability density', colors=colors,
                         show_hist=False, show_curve=True, show_rug=False)
#гистограмма    
    fig.add_traces(list(
        px.histogram(df, x=column_x, color=color, color_discrete_map={'отточные': colors[0], 'не отточные': colors[1]}, 
                       histnorm='probability density',barmode ='overlay').select_traces())
                  )

    fig.update_layout(
    {'title':{'text':'Гистограмма распределения отточных и не отточных клиентов <br> по параметру {}'.format(column_name), 
              'x': 0.5, 'y': 0.95, 'font_size': 20}},
    legend_orientation="h", 
    legend=dict(x=0.2, y=-0.2),#положение легенды
    autosize=False,
    width=900,
    height=500,
    plot_bgcolor='white',
    )
    fig.update_traces(marker_line_color = 'gray', marker_line_width = 1, opacity = 0.8)
    fig.update_xaxes(title_text=column_name, showgrid=True, gridcolor='silver')
    fig.update_yaxes(title_text='плотность вероятности', showgrid=True, gridcolor='silver')

    #fig.show() #показать диаграмму
    fig.show("png") #показать диаграмму для Github
```

#### Найти уровни оттока по категориям.
<a id="analysis3"></a> 

В соответствии с полученными диаграммами распределения непрерывных характеристик, создадим новые категории в каждом параметре, исходя из наибольшего уровня оттока.

**Баланс**

```python
def categorize_balance(row):
    try:
        if 0 <= row <= 724999:
            return 'менее_725000'
        elif 725000 <= row:
            return 'более_725000'
    except:
        return 'ошибка категории'
data['balance_category'] = data['balance'].apply(categorize_balance)
data['balance_category'].value_counts()
```

**Зарплата**

```python
def categorize_salary(row):
    try:
        if 0 <= row <= 99999:
            return 'менее_100000'
        elif 100000 <= row <= 215000:
            return '100000-215000'
        elif 215000 < row:
            return 'более_215000'
    except:
        return 'ошибка категории'
data['salary_category'] = data['salary'].apply(categorize_salary)
data['salary_category'].value_counts()
```

**Возраст**

```python
def categorize_age(row):
    try:
        if 18 <= row <= 24:
            return '18-24'
        elif 25 <= row <= 35:
            return '25-35'
        elif 36 <= row <= 49:
            return '36-49'
        elif 50 <= row <= 60:
            return '50-60'
        elif 61 <= row:
            return '61+'
    except:
        return 'ошибка категории'
data['age_category'] = data['age'].apply(categorize_age)
data['age_category'].value_counts()
```

**Скоринг**

```python
def categorize_score(row):
    try:
        if 0 <= row <= 819:
            return 'низкий'
        elif 820 <= row <= 944:
            return 'средний'
        elif 945 <= row:
            return 'высокий'
    except:
        return 'ошибка категории'
data['score_category'] = data['score'].apply(categorize_score)
data['score_category'].value_counts()
```

**Для проверки гипотезы по количественным независимым данным можно применить 2 метода (параметрический тест - t-test и непараметрический тест Уилкоксона-Манна-Уитни).**

Аспекты, которые надо проверить, чтобы понять, можно ли применять t-тест:

- Выборочные средние должны быть нормально распределены.

Это условие не препятствие. Благодаря ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности. Напомним, это утверждение верно, даже если сама генеральная совокупность не распределена нормально.
Часто это условие путают с необходимостью нормального распределения самих генеральных совокупностей. Для корректного использования t-теста это необязательно.

- Дисперсии рассматриваемых генеральных совокупностей должны быть равны.

Если размеры выборки разного размера или есть сомнения, что дисперсии у совокупностей одинаковые, указывается параметр equal_var как False.

Так как количество данных достаточно велико можно применить t-test. с параметром equal_var= False.

***Проведем тесты сразу по 2 методам.***

**Сформулируем гипотезы.**

H_0: Различий в средней зарплате отточных и не отточных клиентов нет.

H_1: Различия в средней зарплате отточных и не отточных клиентов есть.

**Проверка гипотезы методом `'scipy.stats.ttest_ind'`.**

```python
results = st.ttest_ind(salary_churn_0, salary_churn_1, equal_var=False)# results = вызов метода для проверки гипотезы
print(results.pvalue)# вывод значения p-value на экран 

if results.pvalue < alpha:# условный оператор с выводом строки с ответом
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
results = st.mannwhitneyu(salary_churn_0, salary_churn_1)# results = вызов метода для проверки гипотезы
print(results.pvalue)# вывод значения p-value на экран 

if results.pvalue < alpha:# условный оператор с выводом строки с ответом
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

#### Приоритизировать сегменты
<a id="seg3"></a> 

Создадим таблицу с признаками каждого сегмента, количеством клиентов в сегменте, количеством отточных клиентов, уровнем оттока, средними значениями баланса и зарплаты, а также общей суммой на балансе клиентов по сегменту. 

Приоритизировать сегменты будем по уровню оттока.

```python
def segments_pivot(df, segment_name):

    df1 = [df['user_id'].count(), df['churn'].sum(), df['churn'].mean(), 
           df['balance'].sum(), df['balance'].mean(), df['salary'].mean()]

    df2 = pd.DataFrame(data=df1, columns=[segment_name], index=[
        'размер сегмента', 
        'размер оттока', 
        'доля оттока', 
        'общий баланс', 
        'cредний баланс', 
        'средняя зарплата']
                  )
    return df2
```

```python
segments_attribute = pd.DataFrame(data=['мужчины / возраст 25-35 лет / количеством баллов имущества 3 и более / активные пользователи банка', 
                         'женщины / возраст 25-35 лет / средний уровень скоринга / количество продуктов банка 2 и более', 
                         'возраст 36-49 лет / баланс более 725000 у.е. / средний уровень скоринга', 
                         'возраст  50-60 лет / баланс более 725000 у.е.'], 
                   columns=['признаки сегмента'], 
                   index=['Сегмент 1', 'Сегмент 2', 'Сегмент 3', 'Сегмент 4'])

segments_pivot = pd.concat([
    segments_pivot(segment_1, 'Сегмент 1'),
    segments_pivot(segment_2, 'Сегмент 2'),
    segments_pivot(segment_3, 'Сегмент 3'),
    segments_pivot(segment_4, 'Сегмент 4'), 
],axis=1).T.sort_values(by=['доля оттока'], ascending=False)
segments_pivot.insert (0, 'признаки сегмента', segments_attribute)
segments_pivot['доля оттока'] = round(segments_pivot['доля оттока']*100, 2)
```
